## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
### GPT conversation with Claude

# gpt_model = "gpt-4.1-mini"
# claude_model = "claude-haiku-4-5"

# gpt_system = "You are a chatbot who is very argumentative; \
# you disagree with anything in the conversation and you challenge everything, in a snarky way."

# claude_system = "You are a very polite, courteous chatbot. You try to agree with \
# everything the other person says, or find common ground. If the other person is argumentative, \
# you try to calm them down and keep chatting."

# gpt_messages = ["Hi there"]
# claude_messages = ["Hi"]



### Llama conversation with GEMINI


llama_model = "groq/llama-3.3-70b-versatile"
gemini_model = "openrouter/google/gemini-2.5-flash-lite"

llama_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

gemini_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

llama_messages = [{"role": "user", "content": "Hi there"}]
gemini_messages = [{"role": "user", "content": "Hi"}]
````

````python
# def call_gpt():
#     messages = [{"role": "system", "content": gpt_system}]
#     for gpt, claude in zip(gpt_messages, claude_messages):
#         messages.append({"role": "assistant", "content": gpt})
#         messages.append({"role": "user", "content": claude})
#     response = openai.chat.completions.create(model=gpt_model, messages=messages)
#     return response.choices[0].message.content


### Code for llama and gemini conversation

def call_llama():
    messages = [{"role": "system", "content": llama_system}]
    
    for llama, gemini in zip(llama_messages, gemini_messages):
        messages.append({"role": "assistant", "content": llama.get ("content")}) # .get is used to avoid error in case of missing content key
        messages.append({"role": "user", "content": gemini.get ("content")})
    
    response = completion(
        model=llama_model, 
        messages=messages,
        api_key="your_groq_api_key_here", # Replace with your actual Groq API key
        max_tokens=80
    )
    return response.choices[0].message.content
````

````python
# def call_claude():
#     messages = [{"role": "system", "content": claude_system}]
#     for gpt, claude_message in zip(gpt_messages, claude_messages):
#         messages.append({"role": "user", "content": gpt})
#         messages.append({"role": "assistant", "content": claude_message})
#     messages.append({"role": "user", "content": gpt_messages[-1]})
#     response = anthropic.chat.completions.create(model=claude_model, messages=messages)
#     return response.choices[0].message.content


### Code for llama and gemini conversation



def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]
    
    for llama, gemini in zip(llama_messages, gemini_messages):
        messages.append({"role": "user", "content": llama.get("content")})
        messages.append({"role": "assistant", "content": gemini.get("content")})
        
    # Llama last angry answer is added to the prompt to gemini to respond to it in a calm way
    messages.append({"role": "user", "content": llama_messages[-1].get("content")})
    
    response = completion(
        model=gemini_model, 
        messages=messages,
        api_key="your_openrouter_api_key_here", # Replace with your actual OpenRouter API key
        max_tokens=80
    )
    return response.choices[0].message.content
````

The most reliable way to do this involves thinking a bit differently about your prompts: just 1 system prompt and 1 user prompt each time, and in the user prompt list the full conversation so far.

```python

{conversation}
```

````python
from litellm import completion
from IPython.display import display, Markdown

llama_model = "groq/llama-3.3-70b-versatile"
gemini_model = "openrouter/google/gemini-2.5-flash-lite"
third_model = "groq/llama-3.1-8b-instant" 

api_key_groq = "your_groq_api_key_here" # Replace with your actual Groq API key
api_key_or = "your_openrouter_api_key_here" # Replace with your actual OpenRouter API key

conversation = """
Llama: Hi there!
Gemini: Hi!
"""

def call_model(model_name, character_name, system_behavior, api_key):
    system_prompt = f"""
    You are {character_name}, {system_behavior}.
    You are in a 3-way conversation with the other characters.
    """
    
    user_prompt = f"""
    You are {character_name}.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as {character_name}. Do not repeat anyone else's lines, just give your next single response.
    """
    
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    
    response = completion(
        model=model_name,
        messages=messages,
        api_key=api_key,
        max_tokens=80
    )
    return response.choices[0].message.content.strip()

for i in range(3):
    llama_reply = call_model(llama_model, "Llama", "who is very argumentative and snarky", api_key_groq)
    conversation += f"\nLlama: {llama_reply}"
    display(Markdown(f"### Llama:\n{llama_reply}\n"))
    
    gemini_reply = call_model(gemini_model, "Gemini", "who is very polite and tries to calm people down", api_key_or)
    conversation += f"\nGemini: {gemini_reply}"
    display(Markdown(f"### Gemini:\n{gemini_reply}\n"))
    
    third_reply = call_model(third_model, "ThirdBot", "who is very logical, neutral, and acts as a mediator", api_key_groq)
    conversation += f"\nThirdBot: {third_reply}"
    display(Markdown(f"### ThirdBot:\n{third_reply}\n"))
````
